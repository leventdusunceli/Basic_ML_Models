```python

from google.colab import drive
drive.mount('/content/drive')
import os 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
os.chdir("/content/drive/MyDrive/Online_Lecture_Notes/Machine_Learning_AZ/Practice_Folder/Car_safety_evaluation")
dataset= pd.read_csv("car_evaluation.csv", header = None)
dataset.head(5)
```

```python
dataset.columns = ["Price","Maintenance", "Doors","People","Boot_Size","Safety_Rating","Acceptability"]
```

```python
from sklearn import preprocessing 
from sklearn.model_selection import train_test_split

le = preprocessing.LabelEncoder()

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

X[:,0] = le.fit_transform(X[:,0])
X[:,1] = le.fit_transform(X[:,1])
X[:,2] = le.fit_transform(X[:,2])
X[:,3] = le.fit_transform(X[:,3])
X[:,4] = le.fit_transform(X[:,4])
X[:,5] = le.fit_transform(X[:,5])
y = le.fit_transform(y)


x_train,x_test,y_train,y_test = train_test_split(X,y,test_size= 0.1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 6)
classifier.fit(x_train,y_train)

score = classifier.score(x_test,y_test)*100
print("Accuracy: %3.f" %score, "%" )
```

```python
y_pred = classifier.predict(x_test)

y_actu = pd.Series(le.inverse_transform(y_test),name= "Actual")
y_predict = pd.Series (le.inverse_transform(y_pred), name="Predicted")
df_conf = pd.crosstab(y_actu,y_predict) 
df_conf
```
